```python
Portfolio_CB = DataFrame_PRICE[["ISIN", "Description", "Final position", "RFQ Price"]]
Portfolio_CB = Portfolio_CB.reset_index(drop=True)
Portfolio_CB["Type"] = Other["Type"]
Abs_position_Value = abs(Portfolio_CB["Final position"]) * Portfolio_CB["RFQ Price"]
Portfolio_CB["Weight"] = Abs_position_Value / Abs_position_Value.sum()
Portfolio_CB = Portfolio_CB.set_index("ISIN")


Historical = Historical.apply(pd.to_numeric, errors="coerce")

#Historical = Historical.dropna()
Historical_logRet = np.log(Historical / Historical.shift(1))
Historical_logRet = Historical_logRet[1:].fillna(0)
#print("NaN:", Historical_logRet.isna().sum())
#print("")

print("Corporate Bond Portfolio:")
display(Portfolio_CB)
print("Check - sum of weights: ", Portfolio_CB["Weight"].sum().round(3))
```

### Three cases:

- Hedging with IEAC
- Hedging with IHYG ETF
- Hedging with both (OLS regression)

```python
CB_Log_Ret = Historical_logRet[Portfolio_CB.index]
Portfolio_CB_LogRet = CB_Log_Ret.dot(Portfolio_CB["Weight"])

# IEAC ETF

ETF["Log Returns IEAA LN"] = np.log(ETF["IEAA LN"] / ETF["IEAA LN"].shift(1))
#ETF.head()

# IHYG ETF

ETF["Log Returns XHYA GY"] = np.log(ETF["XHYA GY"] / ETF["XHYA GY"].shift(1))
ETF = ETF.dropna()

# IEAC - CASE 1

df_Correlation_IEAA_LN = pd.concat([Portfolio_CB_LogRet, ETF["Log Returns IEAA LN"]], axis=1)
df_Correlation_IEAA_LN = df_Correlation_IEAA_LN.rename(columns={"Log Returns IEAA LN": "IEAA LN"})
df_Correlation_IEAA_LN.columns.values[0] = "IEAA LN"
df_Correlation_IEAA_LN = df_Correlation_IEAA_LN.dropna()

# High Yield - CASE 2

df_Correlation_XHYA_GY = pd.concat([Portfolio_CB_LogRet, ETF["Log Returns XHYA GY"]], axis=1)
df_Correlation_XHYA_GY = df_Correlation_XHYA_GY.rename(columns={"Log Returns XHYA GY": "XHYA GY"})
df_Correlation_XHYA_GY.columns.values[0] = "XHYA GY"
df_Correlation_XHYA_GY = df_Correlation_XHYA_GY.dropna()

# High Yield & IEAC - CASE 3
OLS_data = pd.concat([Portfolio_CB_LogRet, ETF["Log Returns IEAA LN"], ETF["Log Returns XHYA GY"]], axis=1).dropna()
OLS_data.columns = ["Portfolio", "IEAA LN", "XHYA GY"]
X = OLS_data[["IEAA LN", "XHYA GY"]]
X = sm.add_constant(X)
y = OLS_data["Portfolio"]

OLS_Regression = sm.OLS(y, X).fit()


#


Correlation_IEAA_LN = ETF["Log Returns IEAA LN"].corr(Portfolio_CB_LogRet)
Correlation_XHYA_GY = ETF["Log Returns XHYA GY"].corr(Portfolio_CB_LogRet)

print("Correlation IEAA LN:     ", Correlation_IEAA_LN.round(3))
print()
print("Correlation XHYA GY:     ", Correlation_XHYA_GY.round(3))
print()
print(OLS_Regression.summary())
```

```python
print("            ------------------------------------------")
print("                 ADF – Augmented Dickey-Fuller        ")
print("            ------------------------------------------")
print("")

for series, name in zip([y, X["IEAA LN"], X["XHYA GY"]], ["Portfolio", "IEAA LN", "XHYA GY"]):
    result = adfuller(series)
    print(f"{name}: p-value = {result[1]:.4f}", "-->", np.where(result[1] < 0.05, f"{name} has a Stationary series!", f"{name} investigate Stationarity!") )
    print("")

print("")
print("                 -------------------------------")
print("                      Breusch–Pagan test        ")
print("                 -------------------------------")
print("")
test = het_breuschpagan(OLS_Regression.resid, OLS_Regression.model.exog)
print("p-value BP:", round(test[1],5), "-->", np.where(test[1] < 0.05, "Presence of Heteroskedasticity!", "Investigate Homoskedasticity!"))
print("")
```

**Stationarity** allows us to use OLS Regression.

The presence of **Heteroskedasticity** requires the use of robust errors.

```python
OLS_Regression_robust = OLS_Regression.get_robustcov_results(cov_type='HC0')
print(OLS_Regression_robust.summary())
```
